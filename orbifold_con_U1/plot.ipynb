```python
import csv
import numpy as np
Re = []
Im = []
with open("../Zij_distribution.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        Re.append(float(row["ReZ"]))
        Im.append(float(row["ImZ"]))
Re = np.array(Re)
Im = np.array(Im)
```

```python
vals = np.concatenate([Re, Im])
mean = np.mean(vals)
std = np.std(vals)
p99 = np.quantile(np.abs(vals), 0.99)
max_abs = np.max(np.abs(vals))
R = max(4 * std, 1.2 * p99)
print(R)
```

```python
import matplotlib.pyplot as plt

x = Re
y = Im
plt.figure(figsize=(5,5))
plt.scatter(x, y, s=2, alpha=0.3, color="tab:blue")
plt.xlabel(r"$\mathrm{Re}(Z_{ij})$")
plt.ylabel(r"$\mathrm{Im}(Z_{ij})$")
plt.title(r"Scatter of $Z_{ij}$ from continuous simulation")

# if R is not None:
#         plt.axvline(-R, color="red", linestyle="--", linewidth=1)
#         plt.axvline(+R, color="red", linestyle="--", linewidth=1)
#         plt.axhline(-R, color="red", linestyle="--", linewidth=1)
#         plt.axhline(+R, color="red", linestyle="--", linewidth=1)
#         plt.text(R, R, r"$R$", color="red")
```
